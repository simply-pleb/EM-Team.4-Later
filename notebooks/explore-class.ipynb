```python
# Assuming df_class is your dataset

# Step 1: Define the threshold for the percentage of zeros
threshold = 0.95  # 95% of the values are zero

# Step 2: Remove columns where the percentage of zeros is greater than the threshold
columns_to_drop = []
for col in df_class_numeric.columns:
    # Calculate the proportion of zero values in the column
    proportion_zero = (df_class_numeric[col] == 0).mean()
    
    # If the proportion is greater than the threshold, mark for removal
    if proportion_zero > threshold:
        columns_to_drop.append(col)

# Step 3: Drop the identified columns from the dataframe
df_class_numeric = df_class_numeric.drop(columns=columns_to_drop)

# Output the columns that were removed
print(f"Columns removed: {columns_to_drop}")
```

### Clustering

```python
# Import necessary libraries for clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Step 1: Preprocessing the Data

# Drop non-numeric or identifier columns, like 'repository'
# Assuming 'repository' is the only non-numeric column
df_clustering = df_class_numeric.copy()

# Handle missing values if there are any (optional)
# For example, we can fill missing values with the column mean
df_clustering.fillna(df_clustering.mean(), inplace=True)

# Standardize the data to have mean=0 and variance=1
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clustering)

# Step 2: Determine the Optimal Number of Clusters using the Elbow Method

# Elbow method to find optimal number of clusters
sse = []  # Sum of squared distances for each number of clusters
k_range = range(2, 11)  # Try between 2 and 10 clusters

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    sse.append(kmeans.inertia_)  # Append the sum of squared distances (inertia)

# Plot the Elbow graph
plt.figure(figsize=(8, 6))
plt.plot(k_range, sse, 'bo-', marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Distances (Inertia)')
plt.show()

# Step 3: Optional - Evaluate Clustering with Silhouette Score

# Compute silhouette scores for different values of k
silhouette_scores = []
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    score = silhouette_score(df_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(8, 6))
plt.plot(k_range, silhouette_scores, 'bo-', marker='o')
plt.title('Silhouette Score for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()

# Step 4: Apply K-Means Clustering with Optimal k

# From the elbow or silhouette method, let's assume the optimal number of clusters is 'optimal_k'
optimal_k = 3  # Change this based on the elbow method or silhouette score

# Apply KMeans with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(df_scaled)

# Add the cluster labels back to the original dataframe
df_clustering['Cluster'] = clusters

# Step 5: Visualizing the Clusters (2D or 3D Visualization)

# Let's visualize clusters using PCA for dimensionality reduction
from sklearn.decomposition import PCA

# Reduce to 2D for easy visualization
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=df_clustering['Cluster'], cmap='viridis', marker='o')
plt.title(f'K-Means Clustering with {optimal_k} Clusters (PCA-reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()

# Optional: 3D Plot for visualization (useful if you want to see more dimensions)
from mpl_toolkits.mplot3d import Axes3D

# Reduce to 3D for visualization
pca_3d = PCA(n_components=3)
df_pca_3d = pca_3d.fit_transform(df_scaled)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(df_pca_3d[:, 0], df_pca_3d[:, 1], df_pca_3d[:, 2], c=df_clustering['Cluster'], cmap='viridis', marker='o')
plt.title(f'K-Means Clustering with {optimal_k} Clusters (3D PCA-reduced)')
ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')
plt.colorbar(sc)
plt.show()

# Step 6: Analyze the Clusters

# Print cluster centers for each cluster
print("Cluster Centers (scaled features):")
print(kmeans.cluster_centers_)

# Group the original dataframe by clusters and compute mean values for each cluster
cluster_analysis = df_clustering.groupby('Cluster').mean()
print(cluster_analysis)
```
