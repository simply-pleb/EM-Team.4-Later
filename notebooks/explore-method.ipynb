```python
import pandas as pd
import numpy as np
```

```python
df_method_full = pd.read_csv("../data/method_metrics_all.csv")
df_method_full
```

```python
df_method_numeric = df_method_full.drop(columns=['repository'])
df_method_numeric.head()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# -------------------------------
# 3. Visualizing Distributions
# -------------------------------

# Plot histograms for each numeric feature
df_method_numeric.hist(bins=20, figsize=(18, 12))
plt.tight_layout()
plt.show()

# Plot boxplots to identify outliers
plt.figure(figsize=(18, 12))
sns.boxplot(data=df_method_numeric)
plt.xticks(rotation=90)
# plt.title('Boxplot for All Features')
plt.show()
```

```python
# Assuming df_class is your dataset

# Step 1: Define the threshold for the percentage of zeros
threshold = 0.95  # 95% of the values are zero

# Step 2: Remove columns where the percentage of zeros is greater than the threshold
columns_to_drop = []
for col in df_method_numeric.columns:
    # Calculate the proportion of zero values in the column
    proportion_zero = (df_method_numeric[col] == 0).mean()
    
    # If the proportion is greater than the threshold, mark for removal
    if proportion_zero > threshold:
        columns_to_drop.append(col)

# Step 3: Drop the identified columns from the dataframe
df_method_numeric = df_method_numeric.drop(columns=columns_to_drop)

# Output the columns that were removed
print(f"Columns removed: {columns_to_drop}")
```

```python
from sklearn.preprocessing import StandardScaler

# Assuming df_class is your dataset

# Step 1: Standardize the data
scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df_method_numeric), columns=df_method_numeric.columns)
df_standardized_log = pd.DataFrame(scaler.fit_transform(np.log(df_method_numeric)), columns=df_method_numeric.columns)

plt.figure(figsize=(18, 6))

# Box plot for standardized data
plt.subplot(1, 2, 1)
sns.boxplot(data=df_standardized)
plt.title('Box Plot of Standardized Method Metrics')
plt.ylabel('Standardized Value')

# Box plot for log-transformed standardized data
plt.subplot(1, 2, 2)
sns.boxplot(data=df_standardized_log)
plt.title('Box Plot of Log-Transformed Standardized Method Metrics')
plt.ylabel('Log-Transformed Standardized Value')

plt.tight_layout()
plt.show()
```

```python

# -------------------------------
# 4. Correlation Analysis
# -------------------------------

# Compute correlation matrix
correlation_matrix = df_standardized.corr()

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
# plt.title('Correlation Matrix of Software Metrics')
plt.show()
```
